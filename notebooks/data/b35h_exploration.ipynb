### Check image size

Number of distinct image resolutions

```python
dataset_folder = '/content/gdrive/MyDrive/asub_dataset/preprocessed_newest'
folders = os.listdir(dataset_folder)
images = []
image_shapes = []

for folder in folders:
  folder_path = os.path.join(dataset_folder, folder)
  for path in os.listdir(folder_path):
    image_path = os.path.join(os.path.join(dataset_folder, folder), path)
    image = cv2.imread(image_path) 
    images.append(image)
    (height, width) = image.shape[0], image.shape[1]
    image_shapes.append((height, width))
    
unique_shapes = list(set(image_shapes))
len(unique_shapes)
```

### Check image blurriness

```python
def variance_of_laplacian(image):
  return cv2.Laplacian(image, cv2.CV_64F).var()
```

Calculate mean variance of laplacian for images in the dataset

```python
variances = []

for folder in folders:
  folder_path = os.path.join(dataset_folder, folder)
  for path in os.listdir(folder_path):
    image_path = os.path.join(os.path.join(dataset_folder, folder), path)
    image = cv2.imread(image_path)
    variance = variance_of_laplacian(image)
    variances.append(variance)

np.mean(variances)
```
